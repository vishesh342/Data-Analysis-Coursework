```python
#Import the required Libraries
import pandas as pd
import numpy as np 
import random
```

```python
#reading the csv file from defined file path
filepath='/content/drive/MyDrive/Growth Guide Assignment/population_diabetes.csv'
data=pd.read_csv(filepath)
```

```python
#viewing the data heads
data.head()
```

```python
#Separating Dependent and Independent variables
X=data.iloc[:,:-1]
Y=data['Outcome']
```

# 1. Check if all the features are numeric.

```python
#The Following Code iterates through the columns and check their datatypes.Also has a counter to count the total number of numerical features.
numeric_col_count=0
for column in X.columns:
  if(type(column)!='o'):
    print(column+' is Numeric')
    numeric_col_count+=1
  else:
    print(column+' is not numeric')

print('\n')
print('Total Numeric Columns = {}'.format(numeric_col_count))
```

#2. Given dataset contains many (zero) values, replace 0 with NaN value.

```python
#Since there was presence of many 0 values , we replaced it with NaN value
X=X.replace(0,np.nan)
X.head()
```

# 3. Replace NaN with mean , median, mode.

```python
#The following code imputes the the NaN values with the mean values and transform the dataset
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')
head=X.columns
imputer = imputer.fit(X)

X = pd.DataFrame(imputer.transform(X),columns=head)
X.head()
```

## Normal Deviate Z Test

Given population and sample , check sample belong to same population.

H0 - Samples are likely drawn from the same distributions

H1 - Samples are likely drawn from different distributions.


[Know More about ztest](https://microbenotes.com/z-test/)

As a part of common practice it same to assume , the significance value **=0.05**

Let's assume the  'Glucose' feature is normally distributed data.

```python
alpha=0.05

mu=X['Glucose'].mean()
Sig=X['Glucose'].std()
X_bar=X_sample['Glucose'].mean()
n=len(X_sample['Glucose'])
SE=Sig/np.sqrt(n)


#alpha level of 0.05
z_critical=1.96 
```

```python
z_stat=(X_bar-mu)/SE
```

```python
print(z_stat)
```

Since z_stat is less than z_critical we accept the null hypothesis and reject the alternative. Statistically, we say the sample mean is no different than the population mean and thus the sample is drawn from the population.

#One Sample T- Test

Given population and sample , check sample belong to same population.

H0 - Samples are likely drawn from the same distributions

H1 - Samples are likely drawn from different distributions

[Know more about test](https://microbenotes.com/t-test/)

As a part of common practice it same to assume , the significance value **=0.05**


Let's assume the  'Glucose' feature is normally distributed data.

To perform the **t_test** let us take the random sample of **size 30** of **GLUCOSE** feature.

```python
random_sample=np.array(random.sample(list(X_sample['Glucose']),30))
```

```python
tsample=pd.DataFrame(random_sample,columns=['Glucose'])
```

```python
alpha=0.05

pop_mu=X_sample['Glucose'].mean()
x_bar=tsample['Glucose'].mean()
sigma=tsample['Glucose'].std()
N=len(tsample['Glucose'])
sx=sigma/np.sqrt(N)


#alpha level of 0.05
t_critical = 2.093
```

```python
t_stat= (x_bar-pop_mu)/sx
```

```python
print(t_stat)
```

Since t_stat is less than t_critical we accept the null hypothesis and reject the alternative. Statistically, we say the sample mean is no different than the population mean and thus the sample is drawn from the population.
